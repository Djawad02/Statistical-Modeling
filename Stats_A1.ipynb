```python
                                                #DANIA JAWAD
                                                    #CS-B
                                                 #20I-0569
import random as rn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


                                                #QUESTION 1
    
mean = rn.randint(3, 10)   #randomized value of mean
std = 2

print("The initial value of mean: ", mean ," and initial value of Standard deviation is: " ,std) 

#calulating normal distribution
normalDist =  stats.norm(mean, std) 

print("Mean calculated from Normal distribution: ", normalDist.mean() ," and Standard deviation is: " ,normalDist.std()) 

#plotting graphs
x = np.linspace(-6, 6, 100)
y = (1/(std*np.sqrt(2*np.pi))) * np.exp(-(x-std)**2 / (2*std**2))

plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

s = normalDist.rvs(100)
plt.hist(s, color="green", density=True)
plt.title("plot for 100 normal distribution samples")
plt.show()

print("------------------------------------------------------------------------------------------")



val1 =9
val2 =  0.7
mean1 =6.3

#calulating gamma distribution
gamDist = stats.gamma(val1,scale=val2)

print("The initial value of mean: ", mean1 ," and initial value of Standard deviation is: " ,std) 
print("Mean calculated from gamma distribution: ", gamDist.mean() ," and Standard deviation is: " ,gamDist.std()) 

#plotting graphs
x1 = np.linspace(-6, 6, 100)
y1 = (1/(std*np.sqrt(2*np.pi))) * np.exp(-(x1-std)**2 / (2*std**2))

plt.plot(x1, y1)
plt.title('Gamma Distribution')
plt.show()

s1 = gamDist.rvs(100)
plt.hist(s1, color="yellow", density=True)
plt.title("plot for 100 gamma distribution samples")
plt.show()



                                               


```

```python
                                                      #QUESTION 2

import random as rn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

seedlings = 12
prior = 1
germi = 3
alph = prior + germi  
bet = prior + seedlings - germi  

postDist = stats.beta(alph, bet)

dense = 0.94
interv = postDist.interval(dense)
HDI = postDist.interval(dense)
print("equal-tailed:", interv," and highest density interval:", HDI)

sampleSpace = 89090
oneSeed = stats.binom(n=seedlings, p=postDist.rvs(size=sampleSpace))


probOne = 1 - oneSeed.cdf(0)
print("At least one germination has probability of abt :", probOne.mean())

```
